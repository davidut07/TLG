```python
import pandas as pd
from google.colab import files

# Extraer info
uploaded = files.upload()
csv_file = list(uploaded.keys())[0]
data = pd.read_csv(csv_file)

# b. Mostrar la estructura y tipos de datos de cada columna
print("A continuación se muestra la estructura del DataFrame:")
print(data.info())

print()

# Columna data a datatime64
data['date'] = pd.to_datetime(data['date'])

# c. Determinar la cantidad de vacunas aplicadas de cada compañía
conteo_vacunas = data['vaccines'].value_counts().to_string()
print("Cantidad de vacunas aplicadas por compañía:")
print(conteo_vacunas)

print()

# d. Vacunas aplicadas en todo el mundo
total_vacunas = data['total_vaccinations'].sum()
print("Cantidad de vacunas aplicadas en todo el mundo:", '{:,.0f}'.format(total_vacunas))

print()

# e. Vacunas aplicadas por país
vacunas_pais = data.groupby('country')['total_vaccinations'].mean().map('{:,.0f}'.format).to_string()
print("Promedio de vacunas aplicadas por país:")
print(vacunas_pais)

print()

# f. Vvacunas aplicadas el día 29/01/21 en todo el mundo
total_290121 = data[data['date'] == '2021-01-29']['total_vaccinations'].sum()
print("Cantidad de vacunas aplicadas el 29/01/21 en todo el mundo:", '{:,.0f}'.format(total_290121))

print()

# g. Crear un dataframe nuevo denominado conDiferencias
data['diferencias'] = data['daily_vaccinations'] - data['daily_vaccinations_raw']
conDiferencias = data.copy()

# h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua
periodo = data['date'].max() - data['date'].min()
print("Período de tiempo entre la fecha más reciente y la fecha más antigua:", periodo)

print()

# i. Crear un dataframe nuevo denominado conCantidad
data['canVac'] = data['vaccines'].str.split('/').str.len()
conCantidad = data.copy()

# j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020
antes20 = data[data['date'] < '2020-12-20']

# k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer
pfizer = data[data['vaccines'].str.contains('Pfizer')]

# l. Almacenar los dataframes en un archivo de Excel
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

# Descargar el archivo de Excel
files.download('resultadosReto.xlsx')



```
